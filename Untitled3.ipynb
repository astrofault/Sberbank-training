```python
train = data.loc[data['REPORTDATE'] < '2019-02-01']
test = data.loc[(data['REPORTDATE'] >= '2019-02-01') & (data['REPORTDATE'] <= '2019-03-04')]

train.shape, test.shape
```

```python
test.sort_values(by = 'REPORTDATE')
```

```python
from fbprophet import Prophet
```

```python
model = Prophet()
model
```

```python
train.columns = ['ds', 'y'] 
test.columns = ['ds', 'y']
```

```python
model.fit(train)
```

```python
prediction_size = 30
future = model.make_future_dataframe(periods=prediction_size)
future.tail(30)
```

```python
forecast = model.predict(future)
forecast
```

```python
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(30)
```

```python
model.plot(forecast);
```

```python
model.plot_components(forecast);
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(forecast['yhat'].tail(30), test['y'])
```

```python
modul_diff = np.abs(forecast.tail(30)['yhat'] - test['y'])
dif = np.mean(modul_diff / test['y'])
dif_percent = dif*100
dif_percent
```

```python
model_linear = Prophet(growth= 'linear', yearly_seasonality='auto', weekly_seasonality='auto', seasonality_mode='multiplicative')
```

```python
model_linear.fit(train)
```

```python
forecast_linear = model_linear.predict(future)
```

```python
modul_diff1 = np.abs(forecast_linear.tail(30)['yhat'] - test['y'])
dif1 = np.mean(modul_diff1 / test['y'])
dif_percent1 = dif*100
dif_percent1
```
